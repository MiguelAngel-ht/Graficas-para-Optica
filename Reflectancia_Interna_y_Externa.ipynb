La reflectancia es el cuadrado del coeficiente de reflexión. Para el modo Transversal Eléctrico es

$R_E = \Gamma_{TE}^2 = \left(\frac{ \cos\theta_i - n\cos\theta_t   }{n\cos\theta_t + \cos\theta_i } \right)^2$

y para el modo Transversal Magnético es

$ R_M =  \Gamma^2_{TM} = \left(  \frac{ -n \cos\theta_i + \cos\theta_t}{\cos\theta_t + n\cos\theta_i} \right)^2$

Para una **incidencia interna**: $n<0$

y para una **incidencia externa** : $n>0$

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
# GRÁFICA

n1 = 1.5
n2 = 1
n = n2/n1

oi = np.arange(0,np.pi/2,np.pi/300)
ot = np.arcsin((1/n)*np.sin(oi))  
re = (np.cos(oi) - n*np.cos(ot)) / (n*np.cos(ot) + np.cos(oi))
rm = (-n*np.cos(oi) + np.cos(ot)) / (np.cos(ot) + n*np.cos(oi))

Re = re*re
Rm = rm*rm

plt.figure(figsize=(16,9))
plt.plot(oi,Re, label="Reflectancia Elec. Int.")
plt.plot(oi,Rm, label="Reflectancia Magn. Int.")
plt.vlines(x=np.arctan(n), ymin=-0.1,ymax=1,linestyles= "--",label="Ángulo de Brewster", colors="gray")
plt.vlines(x=np.arcsin(n), ymin=-0.1,ymax=1,linestyles= "--",label="Ángulo Crítico" , colors="k")

n = n1/n2

ot = np.arcsin((1/n)*np.sin(oi)*np.pi/180)
re = (np.cos(oi) -n*np.cos(ot)) / (n*np.cos(ot)+np.cos(oi))
rm = (-n*np.cos(oi) +np.cos(ot)) / (np.cos(ot)+n*np.cos(oi))

Re = re*re
Rm = rm*rm

plt.plot(oi,Re, label="Reflectancia Elec. Ext.")
plt.plot(oi,Rm, label="Reflectancia Magn. Ext.")

plt.title("Reflectancia Interna y Externa",fontsize=28,color='black')
plt.xlabel("Ángulo Incidente (rad)",fontsize=20,color='black')
plt.ylabel("Reflectancia",fontsize=20,color='black')

plt.legend()
#plt.savefig("reflectancia interna y externa.png") #Descomentar para Guardar Imagen
plt.show()
```
