#**Índice de Refracción y Coeficiente de Extinción**
*Por: Miguel Ángel Hernández Tapia*

Una expresión para obtener el índice de refracción en función de la frecuencia angular es

$ n_R^2 - n_I^2 =  1 + \frac{N e^2}{ m \epsilon_0} \left[   \frac{\omega_0 ^2 -\omega^2}{(\omega_0 ^2 -\omega^2 )^2 + \omega^2 \gamma^2} \right]$

y el coeficiente de extinción es

$ 2( n_R n_I) =  \frac{N e^2}{ m \epsilon_0} \left[ j\frac{\omega \gamma}{(\omega_0 ^2 -\omega^2 )^2 + \omega^2 \gamma^2}   \right] $

Se dibuja la gráfica para los valores del electrón con $N= 10^{28} $ $1/m^3$, $w_0 = 
10^{16}$ $ 1/s$ y $\gamma = 10^{14}$ $1/s$ y el resultado se muestra en la siguiente gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.arange(80e15,130e15,10e12)
N = 10e28
e = 1.6e-19
w0 = 10e16
g = 10e14
e0 = 8.85e-12
m = 9.1e-31

y1 = 1 + (N * e**2)/(m*e0)*((w0**2 - w**2)/((w0**2-w**2)**2 + w**2*g**2))
y2 = (N * e**2)/(m*e0)*((w*g)/((w0**2-w**2)**2 + w**2*g**2))


plt.figure(figsize=(16,9))
ax = plt.gca()
ax.axvline( x=w0, c='g',ls='--', lw=2, label='Frecuencia de Resonancia')
ax2 = ax.twinx()
ax.plot(w,y1,c="red")
ax2.plot(w,y2,c="blue")
plt.xscale("log")
plt.title("Índice de refracción y Coeficiente de Extinción",fontsize=28, color='black')
ax.set_ylabel("Indice de Refracción",fontsize=20,color='red')
ax2.set_ylabel("Coeficiente de Extinción",fontsize=20,color='blue')
ax.set_xlabel('Frecuencia', fontsize=20, color='black')
ax.legend()
#plt.savefig("PloT indice de ref y coe de extin.png") #descomentar para guardar imagen
plt.show()
```
