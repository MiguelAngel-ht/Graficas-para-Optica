La parte real (índice de refracción) del índice de refracción del modelo de Drude, es

$n_R = \left[ \frac{1}{2} \left( 1-a + \sqrt{(1-a)^2 +b^2 a^2 } \right) \right]^{1/2}$

y la parte imaginaria (coeficiente de extinción) es

$n_I = \frac{ab\sqrt{2}}{2} \left( \frac{1}{n_R} \right)$

donde 

$a = \frac{\omega_p ^2}{\gamma^2 + \omega ^2}$

y

$b = \frac{\gamma}{\omega}$

con $\omega_p$ la frecuencia angular libre del plasma.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gráfica para la Plata

w = np.arange(1e14,1e17,1e13)
r = (4.1)*(1e13)
wp = (1.63)*(1e16) #--Plata  

a = (wp*wp)/(w*w + r*r)
b = (r/w)
nr = np.sqrt((1-a+np.sqrt((1-a)*(1-a) + (a*b)*(a*b)))*0.5)
ni = b*a*(1/nr)*(np.sqrt(2)/2)

plt.figure(figsize=(16,9))
ax = plt.gca()
ax2 = ax.twinx()
ax.plot(w,ni,c="blue",label="Coeficiente de Extinción")
ax2.plot(w,nr,c="red",label="Indice de Refracción")
ax.axvline( x=wp, c='g',ls='--', lw=2, label='Frecuencia del plasma')
plt.xscale("log")
plt.title("Indice de Refracción y Extinción (Plata)",fontsize=28, color='black')
ax.set_ylabel("nI",fontsize=20,color='blue')
ax2.set_ylabel("nR",fontsize=20,color='red')
ax.set_xlabel('Frecuencia angular', fontsize=14, color='black')
ax2.legend(loc = "right")
ax.legend()
#plt.savefig("Indice de Refracción y Extinción (Plata).png")
plt.show()
```

```python
# Gráfica para el Oro

w = np.arange(1e14,1e17,1e13)
r = (4.1)*(1e13)
wp = (1.3)*(1e16)  #--Oro

a = (wp*wp)/(w*w + r*r)
b = (r/w)
nr = np.sqrt((1-a+np.sqrt((1-a)*(1-a) + (a*b)*(a*b)))*0.5)
ni = b*a*(1/nr)*(np.sqrt(2)/2)

plt.figure(figsize=(16,9))
ax = plt.gca()
ax.axvline( x=wp, c='g',ls='--', lw=2, label='Frecuencia')
ax2 = ax.twinx()
ax.plot(w,ni,c="blue",label="Coeficiente de Extinción")
ax2.plot(w,nr,c="red",label="Indice de Refracción")
plt.xscale("log")
plt.title("Indice de Refracción y Extinción (Oro)",fontsize=28, color='black')
ax.set_ylabel("nI",fontsize=20,color='blue')
ax2.set_ylabel("nR",fontsize=20,color='red')
ax.set_xlabel('Frecuencia angular', fontsize=14, color='black')
ax2.legend(loc = "right")
ax.legend()
#plt.savefig("Indice de Refracción y Extinción (Oro).png")
plt.show()
```
