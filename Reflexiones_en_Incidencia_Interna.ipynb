El coeficiente de reflexión para el modo Transversal Eléctrico es

$\Gamma_{TE} =  \frac{ \cos\theta_i - n\cos\theta_t   }{n\cos\theta_t + \cos\theta_i }$

y para el modo Transversal Magnético es

$\Gamma_{TM} =  \frac{ -n \cos\theta_i + \cos\theta_t}{\cos\theta_t + n\cos\theta_i} $

donde $n<0$ para reflexión interna.

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
# GRÁFICA

n1 = 1.52
n2 = 1
n = n2/n1

oc = np.arcsin(n)
oi = np.arange(0,oc,np.pi/300)
ot = np.arcsin((1/n)*np.sin(oi))

re = (np.cos(oi) - n*np.cos(ot)) / (n*np.cos(ot) + np.cos(oi))
rm = (-n*np.cos(oi) + np.cos(ot)) / (np.cos(ot) + n*np.cos(oi))

re = np.append(re,1)
rm = np.append(rm,-1)
oi = np.append(oi,oc)

plt.figure(figsize=(16,9))
plt.plot(oi,re, label="Reflexión Eléctrica")
plt.plot(oi,rm, label="Reflexión Magnética")
plt.vlines(x=np.arctan(n), ymin=-1,ymax=1,linestyles= "--",label="Ángulo de Brewster", colors="gray")
plt.vlines(x=oc, ymin=-1,ymax=1,linestyles= "--",label="Ángulo Crítico" , colors="k")
plt.hlines(y=0,xmin = 0, xmax = 1.6,linestyles= "--", color="silver")

plt.title("Coeficientes de Reflexión en Incidencia Interna",fontsize=28,color='black')
plt.xlabel("Ángulo Incidente (rad)",fontsize=20,color='black')
plt.ylabel("Reflexión",fontsize=20,color='black')
plt.xlim(0, 1.6)
plt.ylim(-1, 1)
plt.legend()
#plt.savefig("grafica reflexion interna re y rm.png") #Guardar gráfica como imagen  
plt.show()
```
