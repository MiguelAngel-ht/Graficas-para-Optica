El cambio de Fase para el modo Transversal Magnético es

$ \phi_{TM}= \left \{ \begin{matrix}  0 & \mbox{si } \theta_i < \theta_B 
\\ \pi & \mbox{si } \theta_B < \theta_B  < \theta_c 
\\ -2\tan^{-1}\left( \frac{\sqrt{\sin^2\theta_i - n^2}}{n^2\cos \theta_i} \right) + \pi &  \mbox{si  } \theta_c  < \theta_i                                 \end{matrix}\right.$

y para el modo Transversal Eléctrico es

$  \phi_{TE}= \left \{ \begin{matrix}  0 & \mbox{si } \theta_i < \theta_c 
\\ 2\tan^{-1}\left(- \frac{\sqrt{\sin^2\theta_i - n^2}}{\cos \theta_i} \right)  &  \mbox{si  } \theta_c  < \theta_i  \end{matrix}\right.$

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
# Gráfica

oi = np.arange(0,np.pi/2,np.pi/200)

n1 = 1
n2 = 1.5
n = n1/n2

OTM = []; OTE = []

ob = np.arctan(n) 
oc =  np.arcsin(n)

for i in oi:
  if(i <= ob):
    OTM.append(0)
    OTE.append(0)
    continue
  if(i <= oc):
    OTM.append(np.pi)
    OTE.append(0)
    continue
  if(i > oc):
    OTM.append(-2*np.arctan(np.sqrt(np.sin(i)*np.sin(i) - n*n )/(n*n*np.cos(i))) + np.pi) # + np.pi
    OTE.append(2*np.arctan(-(np.sqrt(np.sin(i)*np.sin(i) - n*n )/(np.cos(i)))))
    continue

plt.figure(figsize=(16,9))
plt.plot(oi,OTM, label="Ángulo TM")
plt.plot(oi,OTE, label="Ángulo TE")

plt.title("Cambio de Fase por Ángulo de Incidencia",fontsize=28,color='black')
plt.xlabel("Ángulo Incidente (rad)",fontsize=20,color='black')
plt.ylabel("Cambio de Fase (rad)",fontsize=20,color='black')
plt.legend()
#plt.savefig("Cambio de fase.png")
plt.show()
```
