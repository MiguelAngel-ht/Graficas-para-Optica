El coeficiente de reflexión para el modo Transversal Eléctico es

$ \Gamma_{TE} =  \frac{ \cos\theta_i - n\cos\theta_t   }{n\cos\theta_t + \cos\theta_i }$

y para el modo Transversal Magnético es

$\Gamma_{TM} = \frac{ -n \cos\theta_i + \cos\theta_t}{\cos\theta_t + n\cos\theta_i}$

El coeficiente de transmisión para el modo Traversal Eléctrico es

$\tau_{TE} = \frac{2cos\theta_i  }{n \cos\theta_t +   \cos\theta_i}$

y para el modo Transversal Magnético es

$\tau_{TM} = \frac{2\cos\theta_i }{\cos\theta_t+ n  \cos\theta_i}$

donde $n = n_2/n_1$.

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
# GRÁFICA 

n1 = 1
n2 = 1.62
n = n2/n1

oi = np.arange(0,np.pi/2,np.pi/90)
ot = np.arcsin((n1/n2)*np.sin(oi)*(np.pi/180))

re = (np.cos(oi) -n*np.cos(ot)) / (n*np.cos(ot)+np.cos(oi))
te = 2*np.cos(oi)/(n*np.cos(ot)+np.cos(oi))
rm = (-n*np.cos(oi) +np.cos(ot)) / (np.cos(ot)+n*np.cos(oi))
tm = 2*np.cos(oi)/(np.cos(ot)+n*np.cos(oi))

plt.figure(figsize=(16,9))
plt.plot(oi,re, label="Reflexión Eléctrica")
plt.plot(oi,te, label="Transmisión Eléctrica")
plt.plot(oi,rm, label="Reflexión Magnética")
plt.plot(oi,tm, label="Transmisión Magnética")

plt.title("Reflexión y Transmisión Caso Elécrico y Magnético",fontsize=28,color='black')
plt.xlabel("Ángulo Incidente (rad)",fontsize=20,color='black')
plt.ylabel("Coeficiente",fontsize=20,color='black')
plt.legend()
#plt.savefig("ryt caso electrico y magnetico.png") # Descomentar para Guardar Imagen
plt.show()
```
