Los datos para el dialéctrico corresponden al vidrio y para el conductor es el cobre.

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
# Dialéctrico Con Pérdidas

f = np.arange(0.5e14,2e16,0.1e14) # ESPECTRO VISIBLE

er = 10        
mr = 1           
s = 10e-10      

e0 = 8.85e-12           
m0 = 1.26e-6               

e = e0*er
m = m0*mr

w = 2*np.pi*f 

a1 = []*len(w); b1 = []*len(w); n1 = []*len(w)

for i in w:
  a1.append( i*np.sqrt((m*e/2)*(np.sqrt(1+(s*s/(i*i*e*e)))-1)))
for i in w:
  b1.append( i*np.sqrt((m*e/2)*(np.sqrt(1+(s*s/(i*i*e*e)))+1)) )
for i in w:
  n1.append(abs(np.sqrt(1j*i*m/(s+1j*i*e)) ))
```

```python
# Dialéctrico Sin Pérdidas

er = 10       
mr = 1           
s = 10e-10     

e = e0*er
m = m0*mr

a2 = []*len(w); b2 = []*len(w); n2 = []*len(w)

for i in w:
  a2.append(0)
for i in w:
  b2.append( i*np.sqrt(m*e))
for i in w:
  n2.append(np.sqrt(m/e))
```

```python
# Vacío 

e = e0
m = m0

a3 = []*len(w); b3 = []*len(w); n3 = []*len(w)

for i in w:
  a3.append(0)
for i in w:
  b3.append( i*np.sqrt(m*e))
for i in w:
  n3.append(np.sqrt(m/e))
```

```python
# Conductor 
                                  
er = 1                          
mr = 1                            
s = 5.80e7                        

e = e0*er
m = m0*mr

a4 = []*len(w); b4 = []*len(w); n4 = []*len(w); k=0

for i in w:
  a4.append( np.sqrt(i*m*s/s) )
for i in w:
  b4.append( np.sqrt(i*m*s/s)  )
for i in w:
  n4.append(abs((a4[k]/s )*(1+1j)))
  k = k+1
```

```python
# GRÁFICAS

fig2 = plt.figure(constrained_layout=True)
spec2 = gridspec.GridSpec(ncols=4, nrows=3, figure=fig2)
fig2 = plt.figure(figsize=(25,11))

f2_ax1 = fig2.add_subplot(spec2[0, 0])
f2_ax2 = fig2.add_subplot(spec2[0, 1])
f2_ax3 = fig2.add_subplot(spec2[0, 2])
f2_ax4 = fig2.add_subplot(spec2[0, 3])

f2_ax1.plot(w,a1 ,c="g")
f2_ax2.plot(w,a2 ,c="b")
f2_ax3.plot(w,a3 ,c="r")
f2_ax4.plot(w,a4 ,c="k")

f2_ax1.set_title('Dialéctrico con pérdidas',fontsize=16)
f2_ax2.set_title('Dialéctrico sin pérdidas',fontsize=16)
f2_ax3.set_title('Vacío',fontsize=16)
f2_ax4.set_title('Conductor',fontsize=16)
f2_ax1.set_ylabel("α",fontsize=20)
f2_ax1.set_xscale('log') ; f2_ax1.set_yscale('linear') 
f2_ax2.set_xscale('log')
f2_ax3.set_xscale('log')
f2_ax4.set_xscale('log') ; f2_ax4.set_yscale('linear') 
f2_ax5 = fig2.add_subplot(spec2[1, 0])
f2_ax6 = fig2.add_subplot(spec2[1, 1])
f2_ax7 = fig2.add_subplot(spec2[1, 2])
f2_ax8 = fig2.add_subplot(spec2[1, 3])


f2_ax5.set_xscale('log') ; f2_ax5.set_yscale('linear') 
f2_ax6.set_xscale('log') ; f2_ax6.set_yscale('linear') 
f2_ax7.set_xscale('log') ; f2_ax7.set_yscale('linear') 
f2_ax8.set_xscale('log') ; f2_ax8.set_yscale('linear') 
f2_ax5.plot(w,b1 ,c="g")
f2_ax5.set_ylabel("β",fontsize=20)
f2_ax6.plot(w,b2 ,c="b")
f2_ax7.plot(w,b3 ,c="r")
f2_ax8.plot(w,b4 ,c="k")

f2_ax9 = fig2.add_subplot(spec2[2, 0])
f2_ax10 = fig2.add_subplot(spec2[2, 1])
f2_ax11 = fig2.add_subplot(spec2[2, 2])
f2_ax12 = fig2.add_subplot(spec2[2, 3])



f2_ax9.set_xscale('log') ; f2_ax9.set_yscale('linear') 
f2_ax10.set_xscale('log') 
f2_ax11.set_xscale('log')  
f2_ax12.set_xscale('log') ; f2_ax12.set_yscale('linear') 
f2_ax9.plot(w,n1 ,c="g")
f2_ax9.set_ylabel("η",fontsize=20)
f2_ax10.plot(w,n2 ,c="b")
f2_ax11.plot(w,n3 ,c="r")
f2_ax12.plot(w,n4 ,c="k")

f2_ax9.set_xlabel("Frequencia")
f2_ax10.set_xlabel("Frequencia")
f2_ax11.set_xlabel("Frequencia")
f2_ax12.set_xlabel("Frequencia")

#plt.savefig("Graficas.png") # Descomentar para Guardar imagen 
plt.show()
```
