La ecuación obtenida para el coeficiente de reflexión es

$ \Gamma_\parallel = \frac{\eta_1 \cos\theta_t - \eta_2 \cos\theta_i  }{\eta_1 \cos\theta_t + \eta_2 \cos\theta_i  }$

y para el coeficiente de transmisión es

$  \tau_\parallel = \frac{2\eta_1\cos\theta_i}{\eta_2 \cos\theta_i + \eta_1\cos\theta_t} $

donde $\eta = n/\mu $.

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
# GRÁFICA

m0 = 4*np.pi*1e-17     # permeabilidad en el vacío
n1 = 1
m1 = m0               
n2 = 1.33
m2 = m0          # para efectos prácticos en el agua
η = n1/m1
η_ = n2/m2

oi = np.arange(0,np.pi/2,np.pi/90)
ot = np.arcsin((n1/n2)*np.sin(oi)*(np.pi/180))

r = (η *np.cos(ot) - η_*np.cos(oi))/(η*np.cos(ot) + η_*np.cos(oi))
t = 2*η*np.cos(oi) / (η_*np.cos(oi) + η*np.cos(ot))

plt.figure(figsize=(16,9))
plt.plot(oi,r,label="Reflexión H2O")
plt.plot(oi,t,label="Transmisión H2O")


n2 = 1.51
m2 = (0.99)*m0        # para Bk7
η_ = n2/m2

r = (η*np.cos(ot) - η_*np.cos(oi))/(η*np.cos(ot) + η_*np.cos(oi))
t = 2*η*np.cos(oi) / (η_*np.cos(oi) + η*np.cos(ot))

plt.plot(oi,r,label="Reflexión BK7")
plt.plot(oi,t,label="Transmisión BK7")

plt.title("Caso Paralelo",fontsize=30,color='black')
plt.xlabel("Ángulo incidente (radianes)",fontsize=20,color='black')
plt.ylabel("Reflexión",fontsize=20,color='black')
plt.legend()

#plt.savefig("coeficientes paralelo.png") #DESCOMENTAR PARA GUARDAR IMAGEN
plt.show()
```
