La ecuación de Cauchy se aproxima como

$n^2 = A'+\frac{B'}{\lambda^2}+\frac{C'}{\lambda^4}$

donde los primeros dos términos son los más determinantes. La ecuación de Sellmeier por su parte, se escribe como

$ n^2 = A +\frac{B_1 \lambda^2}{\lambda^2 - C_1} + \frac{B_2 \lambda^2}{\lambda^2 - C_2}  $

donde generalmente se usa $A = 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# VALORES DE LOS COEFICIENTES USADOS

A =1.5046 ;B = 0.004;A2 = 1.458 ;B2 = 0.00354 ;A3 = 1.522;B3 =  0.0045 
v = 1.0396;c = 0.006 ; v2 =0.696; c2 = 0.0046; v3 = 1.085; c3 = 0.0199;

L=np.arange(0.3,0.9,0.01) # En micro-metros

y1 = A + B/L**2
y2 = A2 + B2/(L**2)
y3 = 1.728 + 0.01342/(L**2)
n1 = np.sqrt(1 + ((v*(L**2))/(L**2 - c)) + ((0.23*(L**2))/(L**2 - 0.02)))
n2 = np.sqrt(1 + (v2*(L**2))/(L**2 - c2) + ((0.41*(L**2))/(L**2 - 0.013))) 
n3 = np.sqrt(1 + (1.621*(L**2))/(L**2 - 0.012) + ((0.256*(L**2))/(L**2 - 0.06)) + ((1.644*(L**2))/(L**2 -147.47)))

plt.figure(figsize=(16,9))
plt.plot(L,y1,'b')
plt.plot(L,y2,'r')
plt.plot(L,y3,'g')

plt.plot(L,n1, "xb--")
plt.plot(L,n2,"xr--")
plt.plot(L,n3,"xg--")
plt.title("Sellmeier (---) vs Cauchy (―)",fontsize=28,color='black')
plt.ylabel("Longitud de Onda (λ)",fontsize=20,color='black')
plt.xlabel("Índice de Refracción (n)",fontsize=20,color='black')

#plt.savefig("Sellmeier vs Cauchy.png")
plt.show()
```
